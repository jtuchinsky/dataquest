## Matplotlib Styles
So far in this course, we've learned about design and data storytelling. In this lesson, we're going to focus our attention on Matplotlib's pre-defined styles. For this case study, we'll use the `fivethirtyeight` style to build this graph.

![Sample Image](./image3.png)

Matplotlib's pre-defined styles change the default visual properties of graphs. Below, we create a line plot using the `ggplot` style. To do that, we import the `matplotlib.style` submodule and then use the `style.use()` function.

```python
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('ggplot')
plt.plot([2,4,6], [10,15,5])

style.use('default')
plt.plot([2,4,6], [10,15,5])
plt.show()
```


````python
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('ggplot')
plt.plot([2,4,6], [10,15,5])
plt.show()

style.use('default')
plt.plot([2,4,6], [10,15,5])
plt.show()


````

We can see all the available styles by accessing the `style.available` attribute. Let's print out the available styles.

```python
print(style.available)
```
If you want to switch between different styles, use `style.use('default')` between each change — some of the styles can interfere with one another.
